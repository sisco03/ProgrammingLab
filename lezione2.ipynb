```python
import numpy as np

#1.Indicizzazione booleana

#1)
def is_prime(x):

    divisori = 0
    
    for i in range(1,x+1):
        if(x%i==0):
            divisori+=1
    
    if(divisori==2):
        return True
    
    return False

array_primi = np.array([n for n in range(0,21) if is_prime(n)])

#2)
a = array_primi[array_primi > 10]

#3)
primi_pari = array_primi[array_primi % 2 == 0]

#risultati
array_primi, a, primi_pari
```

```python
#2.Operazioni su array

a = np.arange(1,21).reshape(5,4)
#prendo solo la seconda e quarta riga
b = a[[1, 3], :]
#prendo la terza riga
c = a[2, :]

d = a/c

#risultati
a, b, c, d
```

```python
#3.Seleziona il numero più vicino a 0.5 in un array casuale

#matrice 10x3 con valori casuali
random_arr =np.random.rand(10,3)
#uso il valore assoluto e trovo la distanza di ogni elemento da 0.5, poi seleziono la distanza minima per ogni riga
indici_distanza_minima = np.abs(random_arr-0.5).argmin(axis=1)

righe_arr = np.arange(len(random_arr))
#con lo slicing prendo il valore che corrisponde all'indice della distanza minima
valori_distanza_minima = random_arr[righe_arr, indici_distanza_minima]
#risultati
random_arr, indici_distanza_minima, valori_distanza_minima
```

```python
#4.Analisi della frequenza cardiaca con numpy

#1)
bpm = np.array([68, 65, 77, 110, 160, 161, 162, 161, 160, 161, 162, 163, 164, 163, 162, 100, 90, 97, 72, 60, 70])

#2)
bpm_min = np.min(bpm)

#3)
bpm_max = np.max(bpm)

#4)
bpm_sopra_120 = bpm > 120

def percentuale_osservazioni(array):

    cont = 0

    for n in array:
        if n > 120:
            cont += 1
    
    return cont/len(array) * 100

osservazioni = percentuale_osservazioni(bpm)

#risultati
bpm, bpm_min, bpm_max, bpm_sopra_120, osservazioni
```

```python
#5.Manipolazione di vettori

#1)
stipendi = np.array([50000, 105250, 55000, 89000])

costo_tot_iniziale = sum(stipendi)

#2)
#aumento stipendio CEO
stipendi = np.where(stipendi == 105250, 105250*1.15 , stipendi)

#3)
#il valore 121037,50 è presente nell'array dopo aver avuto l'aumento

#4)
stipendi = np.array([50000, 105250, 55000, 89000], dtype=float)

stipendi = np.where(stipendi == 105250, 105250*1.15 , stipendi)

#5)
#aumento per l'impiegata che guadagna di meno
stipendi = np.where(stipendi == 50000, 50000*1.20 , stipendi)

#6)
#aumento per gli altri 2 impiegati rimanenti
stipendi = np.where(stipendi == 55000, 55000*1.10 , stipendi)
stipendi = np.where(stipendi == 89000, 89000*1.10 , stipendi)

#7)
costo_tot_finale = sum(stipendi) - costo_tot_iniziale

#risultati
stipendi, costo_tot_iniziale, costo_tot_finale
```

```python
#6.Valori di Co2

#1)
#carico i dati del file
emissioni = np.loadtxt('emissioni.txt', dtype=float)

#2)
#contatore per segnare i valori >= 18
cont = 0

for value in emissioni:

    if value >= 18:
        cont +=1

media_valori = sum(emissioni)/len(emissioni)

#3)
somma_tot_sopra_la_soglia = 0

for value in emissioni:

    if value >= 18:
        somma_tot_sopra_la_soglia += value

#frazione totale di emissione sopra la soglia
frazione_totale_sopra_la_soglia = somma_tot_sopra_la_soglia/sum(emissioni)

#risultati
emissioni, cont, media_valori, frazione_totale_sopra_la_soglia
```
